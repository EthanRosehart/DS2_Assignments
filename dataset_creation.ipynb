```python
import pandas as pd
import numpy as np
```

```python
emotion_columns = [
    'Answer.f1.afraid.raw', 'Answer.f1.angry.raw', 'Answer.f1.anxious.raw',
    'Answer.f1.ashamed.raw', 'Answer.f1.awkward.raw', 'Answer.f1.bored.raw',
    'Answer.f1.calm.raw', 'Answer.f1.confused.raw', 'Answer.f1.disgusted.raw',
    'Answer.f1.excited.raw', 'Answer.f1.frustrated.raw', 'Answer.f1.happy.raw',
    'Answer.f1.jealous.raw', 'Answer.f1.nostalgic.raw', 'Answer.f1.proud.raw',
    'Answer.f1.sad.raw', 'Answer.f1.satisfied.raw', 'Answer.f1.surprised.raw'
]

# Calculate the count for each emotion
emotion_counts = df_journals[emotion_columns].sum()

# Convert to a DataFrame for easier viewing
emotion_counts_df = emotion_counts.reset_index()
emotion_counts_df.columns = ['Emotion', 'Count']

# Display the result
print(emotion_counts_df)
```

```python
# Consolidating counts for each core emotion
core_emotions = {
    "Joy": [
        'Answer.f1.happy.raw', 'Answer.f1.satisfied.raw',
        'Answer.f1.proud.raw', 'Answer.f1.excited.raw', 'Answer.f1.calm.raw'
    ],
    "Disgust": [
        'Answer.f1.disgusted.raw', 'Answer.f1.jealous.raw',
        'Answer.f1.ashamed.raw'        
    ],
    "Anger": [
        'Answer.f1.frustrated.raw', 'Answer.f1.angry.raw'
        
    ],
    "Fear": [
        'Answer.f1.anxious.raw', 'Answer.f1.afraid.raw',
        'Answer.f1.awkward.raw', 'Answer.f1.confused.raw',
        'Answer.f1.surprised.raw'
    ],
    "Sadness": [
        'Answer.f1.sad.raw', 'Answer.f1.bored.raw',  
        'Answer.f1.nostalgic.raw'
    ]
}

consolidated_counts = {
    emotion: df_journals[columns].sum() for emotion, columns in core_emotions.items()
}

# Display results
for core_emotion, count in consolidated_counts.items():
    print(f"{core_emotion}: {count}")
```

```python
# Create a new DataFrame with journal entries and core emotion classifications
df_core_emotions = df_journals[['Answer']].copy()

# Add a column for each core emotion
for emotion, columns in core_emotions.items():
    # Check if any of the columns for the current emotion are True
    df_core_emotions[emotion] = df_journals[columns].any(axis=1)

# Display the new DataFrame
print(df_core_emotions)
```

```python
# Count the frequency of each core emotion
emotion_frequencies = df_core_emotions[['Joy', 'Disgust', 'Anger', 'Fear', 'Sadness']].sum()

# Convert to a DataFrame for easier readability
emotion_frequencies_df = emotion_frequencies.reset_index()
emotion_frequencies_df.columns = ['Emotion', 'Frequency']

# Display the result
print(emotion_frequencies_df)
sum(emotion_frequencies_df['Frequency'])
```

```python
# Create a column representing the combination of emotions for each row
df_core_emotions['Emotion_Combination'] = df_core_emotions[['Joy', 'Disgust', 'Anger', 'Fear', 'Sadness']].apply(
    lambda row: ', '.join(row.index[row]), axis=1
)

# Count the frequency of each combination
combination_frequencies = df_core_emotions['Emotion_Combination'].value_counts()

# Convert to a DataFrame for better readability
combination_frequencies_df = combination_frequencies.reset_index()
combination_frequencies_df.columns = ['Emotion_Combination', 'Frequency']

# Display the result
print(combination_frequencies_df)
```

```python
df_mentalhealth = pd.read_csv('https://media.githubusercontent.com/media/EthanRosehart/DS2_Assignments/refs/heads/main/students_mental_health.csv')
```

```python
# Filter the mental health dataset
columns_to_keep = ['Age', 'Gender', 'CGPA', 'Stress_Level', 'Depression_Score',
                   'Anxiety_Score', 'Sleep_Quality', 'Physical_Activity', 'Diet_Quality',
                   'Relationship_Status', 'Financial_Stress']
df_mentalhealth_filtered = df_mentalhealth[columns_to_keep].copy()

# Add a Negative_Emotion indicator in df_core_emotions
negative_emotions = ['Disgust', 'Anger', 'Fear', 'Sadness']
df_core_emotions['Negative_Emotion'] = df_core_emotions[negative_emotions].any(axis=1)
```

```python
df_core_emotions['Negative_Emotion'].value_counts()
```

```python
# Filter journal entries with and without negative emotions
negative_emotions = df_core_emotions[df_core_emotions['Negative_Emotion']]
positive_emotions = df_core_emotions[~df_core_emotions['Negative_Emotion']]

# Remove entries where 'Age' > 35 from df_mentalhealth_filtered
df_mentalhealth_filtered = df_mentalhealth_filtered[df_mentalhealth_filtered['Age'] <= 27]

# Reduce 'Age' by 8 for all entries in df_mentalhealth_filtered
df_mentalhealth_filtered['Age'] = df_mentalhealth_filtered['Age'] - 8

# Remove any entries where 'Relationship_Status' == 'Married' from df_mentalhealth_filtered
df_mentalhealth_filtered = df_mentalhealth_filtered[df_mentalhealth_filtered['Relationship_Status'] != 'Married']

# Filter mental health dataset into higher and lower scores
higher_scores = df_mentalhealth_filtered[
    (df_mentalhealth_filtered['Depression_Score'] >= 3) & 
    (df_mentalhealth_filtered['Anxiety_Score'] >= 3)
]
lower_scores = df_mentalhealth_filtered[
    (df_mentalhealth_filtered['Depression_Score'] < 3) & 
    (df_mentalhealth_filtered['Anxiety_Score'] < 3)
]

# Define the necessary columns to check for NULL values
necessary_columns = ['Age', 'Gender', 'CGPA', 'Stress_Level', 'Depression_Score',
                     'Anxiety_Score', 'Sleep_Quality', 'Physical_Activity', 'Diet_Quality',
                     'Relationship_Status', 'Financial_Stress']

# Create filtered versions of higher_scores and lower_scores without NULL values
higher_scores_filtered = higher_scores.dropna(subset=necessary_columns)
lower_scores_filtered = lower_scores.dropna(subset=necessary_columns)

# Ensure that the number of samples does not exceed the available data
num_negative = min(len(negative_emotions), len(higher_scores_filtered))
num_positive = min(len(positive_emotions), len(lower_scores_filtered))

# Sample from higher_scores_filtered for journal entries with negative emotions
sampled_higher_scores = higher_scores_filtered.sample(n=num_negative, random_state=42)

# Sample from lower_scores_filtered for journal entries without negative emotions
sampled_lower_scores = lower_scores_filtered.sample(n=num_positive, random_state=42)

# Introduce randomness: Swap some entries between higher and lower scores
swap_count = int(num_negative * 0.1)  # 10% randomness
swap_count = min(swap_count, len(sampled_higher_scores), len(sampled_lower_scores))

random_higher_to_lower = sampled_higher_scores.sample(n=swap_count, random_state=42)
random_lower_to_higher = sampled_lower_scores.sample(n=swap_count, random_state=42)

# Perform the swaps
sampled_higher_scores = pd.concat([
    sampled_higher_scores.drop(random_higher_to_lower.index),
    random_lower_to_higher
], ignore_index=True)

sampled_lower_scores = pd.concat([
    sampled_lower_scores.drop(random_lower_to_higher.index),
    random_higher_to_lower
], ignore_index=True)

# Combine journal entries with their corresponding scores
negative_combined = pd.concat([
    negative_emotions.reset_index(drop=True),
    sampled_higher_scores.reset_index(drop=True)
], axis=1)

positive_combined = pd.concat([
    positive_emotions.reset_index(drop=True),
    sampled_lower_scores.reset_index(drop=True)
], axis=1)

# Combine all entries into the final dataset
df_final = pd.concat([negative_combined, positive_combined]).reset_index(drop=True)

# Retain necessary columns and drop rows with any NULL values
final_columns = ['Answer', 'Age', 'Gender', 'CGPA', 'Stress_Level', 'Depression_Score',
                 'Anxiety_Score', 'Sleep_Quality', 'Physical_Activity', 'Diet_Quality',
                 'Relationship_Status', 'Financial_Stress', 'Negative_Emotion']
df_final = df_final[final_columns].dropna()

# Display the final DataFrame
print(df_final)
```

```python
# Generate a unique 9-digit Student ID for each entry using permutation
np.random.seed(42)  # For reproducibility
student_ids = np.random.permutation(range(100000000, 100000000 + len(df_final)))
df_final['Student_ID'] = student_ids

# Rearrange columns to place Student_ID at the front
columns_order = ['Student_ID'] + [col for col in df_final.columns if col != 'Student_ID']
df_final = df_final[columns_order]

```

```python
# Function to generate absences and lates with more randomness
def generate_absences_and_lates(row):
    # Determine if the student is at risk
    at_risk = ((row['Depression_Score'] + row['Anxiety_Score']) / 2) >= 3

    # Base parameters for absences and lates
    # Adjust these values based on realistic data
    base_absence_mean = 5
    base_absence_std = 2
    base_late_mean = 3
    base_late_std = 1.5

    # Introduce overlap by setting similar ranges for both groups
    if at_risk:
        # At-risk students might have slightly higher means
        absence_mean = base_absence_mean + np.random.normal(1, 0.5)
        late_mean = base_late_mean + np.random.normal(0.5, 0.3)
    else:
        # Not at-risk students might have slightly lower means
        absence_mean = base_absence_mean + np.random.normal(-1, 0.5)
        late_mean = base_late_mean + np.random.normal(-0.5, 0.3)

    # Generate absences and lates using normal distributions
    absences = np.random.normal(absence_mean, base_absence_std)
    lates = np.random.normal(late_mean, base_late_std)

    # Ensure absences and lates are non-negative integers
    absences = max(int(round(absences)), 0)
    lates = max(int(round(lates)), 0)

    return absences, lates

# Apply the function to generate absences and lates
df_final[['Absences', 'Lates']] = df_final.apply(
    lambda row: pd.Series(generate_absences_and_lates(row)), axis=1
)

# Add At Risk column
df_final['At_Risk'] = ((df_final['Depression_Score'] + df_final['Anxiety_Score']) / 2 >= 3).astype(int)

# Export to CSV
output_path = "final_dataset.csv"
df_final.to_csv(output_path, index=False)

print(f"Dataset exported to {output_path}")
```
